```python
import itertools
%matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

baris = 4
kolom = 4
jumlah_data = baris * kolom
line = [None] * (jumlah_data)

fig, ax = plt.subplots(baris, kolom, figsize=(15, 10))
ax = ax.ravel()
noUrut = 1
for i in range(jumlah_data):
    title = "Sinyal EMG ch-"+str(noUrut)
    line[i], = ax[i].plot([], [], lw=2)
    ax[i].grid()
    ax[i].set_title(title)
    noUrut = noUrut + 1

xdata1, ydata1 = [], []
xdata2, ydata2 = [], []
xdata3, ydata3 = [], []
xdata4, ydata4 = [], []
xdata5, ydata5 = [], []
xdata6, ydata6 = [], []
xdata7, ydata7 = [], []
xdata8, ydata8 = [], []
xdata9, ydata9 = [], []
xdata10, ydata10 = [], []
xdata11, ydata11 = [], []
xdata12, ydata12 = [], []
xdata13, ydata13 = [], []
xdata14, ydata14 = [], []
xdata15, ydata15 = [], []
xdata16, ydata16 = [], []

def data_gen():
    for cnt in itertools.count():    
        yield (cnt, emg[cnt,0]), (cnt, emg[cnt,1]), (cnt, emg[cnt,2]), (cnt, emg[cnt,3]), (cnt, emg[cnt,4]), (cnt, emg[cnt,5]), (cnt, emg[cnt,6]), (cnt, emg[cnt,7]), (cnt, emg[cnt,8]), (cnt, emg[cnt,9]) , (cnt, emg[cnt,10]), (cnt, emg[cnt,11]), (cnt, emg[cnt,12]), (cnt, emg[cnt,13]), (cnt, emg[cnt,14]), (cnt, emg[cnt,15])
        
def init():
    for i in range(jumlah_data):
        ax[i].set_ylim(-10, 10)
        ax[i].set_xlim(0, 1000)  

        del xdata1[:]
        del ydata1[:]
        line[i].set_data(xdata1, ydata1)
        
    
    return line[0], line[1], line[2], line[3], line[4], line[5], line[6], line[7], line[8], line[9], line[10], line[11], line[12], line[13], line[14], line[15]#, line[16]

def run(data):
    # update the data
    t, y = data[0]
    xdata1.append(t)
    ydata1.append(y)
    
    t, y = data[1]
    xdata2.append(t)
    ydata2.append(y)
    
    t, y = data[2]
    xdata3.append(t)
    ydata3.append(y)
    
    t, y = data[3]
    xdata4.append(t)
    ydata4.append(y)

    t, y = data[4]
    xdata5.append(t)
    ydata5.append(y)
    
    t, y = data[5]
    xdata6.append(t)
    ydata6.append(y)
    
    t, y = data[6]
    xdata7.append(t)
    ydata7.append(y)
    
    t, y = data[7]
    xdata8.append(t)
    ydata8.append(y)
    
    t, y = data[8]
    xdata9.append(t)
    ydata9.append(y)
    
    t, y = data[9]
    xdata10.append(t)
    ydata10.append(y)
    
    t, y = data[10]
    xdata11.append(t)
    ydata11.append(y)
    
    t, y = data[11]
    xdata12.append(t)
    ydata12.append(y)

    t, y = data[12]
    xdata13.append(t)
    ydata13.append(y)
    
    t, y = data[13]
    xdata14.append(t)
    ydata14.append(y)
    
    t, y = data[14]
    xdata15.append(t)
    ydata15.append(y)
    
    t, y = data[15]
    xdata16.append(t)
    ydata16.append(y)
    
    for i in range(jumlah_data):
        xmin, xmax = ax[i].get_xlim()
        ymin, ymax = ax[i].get_ylim()
        
        if t >= xmax:
            ax[i].set_xlim(xmin, 2*xmax)            
            ax[i].figure.canvas.draw()                 
       
    line[0].set_data(xdata1, ydata1) 
    line[1].set_data(xdata2, ydata2) 
    line[2].set_data(xdata3, ydata3) 
    line[3].set_data(xdata4, ydata4) 
    line[4].set_data(xdata5, ydata5) 
    line[5].set_data(xdata6, ydata6) 
    line[6].set_data(xdata7, ydata7) 
    line[7].set_data(xdata8, ydata8) 
    line[8].set_data(xdata9, ydata9) 
    line[9].set_data(xdata10, ydata10) 
    line[10].set_data(xdata11, ydata11) 
    line[11].set_data(xdata12, ydata12) 
    line[12].set_data(xdata13, ydata13) 
    line[13].set_data(xdata14, ydata14) 
    line[14].set_data(xdata15, ydata15) 
    line[15].set_data(xdata16, ydata16) 
        
    return line[0], line[1], line[2], line[3], line[4], line[5], \
           line[6], line[7], line[8], line[9], line[10], line[11], \
           line[12], line[13], line[14], line[15]

ani = animation.FuncAnimation(fig, run, data_gen, init_func=init, interval=0,  cache_frame_data=True, blit=True, save_count=0)
fig.tight_layout()
fig.canvas.manager.full_screen_toggle() # toggle fullscreen mode
plt.show()
```
